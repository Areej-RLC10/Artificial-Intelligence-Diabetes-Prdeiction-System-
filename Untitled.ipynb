```python
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = X.select_dtypes(include=['object']).columns

# Fill missing numeric values with mean
X[numeric_cols] = X[numeric_cols].fillna(X[numeric_cols].mean())

# Encode categorical variables (one-hot encoding)
X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

print("Cleaned X shape:", X.shape)
```

```python
# Convert target to numbers
y_train = y_train.map({'No':0,'Yes':1})
y_test = y_test.map({'No':0,'Yes':1})

models = {
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Random Forest": RandomForestClassifier(n_estimators=150, random_state=42),
    "SVM": SVC(probability=True, random_state=42) }
for name, model in models.items():
    
    print("\n==============================")
    print(f" Model: {name}")
    print("==============================")
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy:.4f}")
    
    if hasattr(model, "predict_proba"):
        probs = model.predict_proba(X_test)[:, 1]
    else:
        probs = model.decision_function(X_test)
    
    fpr, tpr, _ = roc_curve(y_test, probs)
    roc_auc = auc(fpr, tpr)
    print(f"AUC Score: {roc_auc:.4f}")
    
    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
    plt.plot([0,1], [0,1], '--')
    plt.title(f"{name} - ROC Curve")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.show()
```

```python
# Step 1: Create results dictionary
results = {}

# Step 2: Train all models and store metrics
models = {
    "Logistic Regression": LogisticRegression(max_iter=500),
    "Random Forest": RandomForestClassifier(n_estimators=150, random_state=42),
    "SVM": SVC(probability=True, random_state=42)
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # AUC
    if hasattr(model, "predict_proba"):
        probs = model.predict_proba(X_test)[:, 1]
    else:
        probs = model.decision_function(X_test)
    
    fpr, tpr, _ = roc_curve(y_test, probs)
    roc_auc = auc(fpr, tpr)
    
    # Store in results
    results[name] = {"Accuracy": accuracy, "AUC": roc_auc}
```

```python
model_names = list(results.keys())
accuracies = [metrics["Accuracy"] for metrics in results.values()]
aucs = [metrics["AUC"] for metrics in results.values()]

x = range(len(model_names))
plt.figure(figsize=(8,5))
plt.bar(x, accuracies, width=0.4, label="Accuracy", color='skyblue')
plt.bar([i + 0.4 for i in x], aucs, width=0.4, label="AUC", color='salmon')
plt.xticks([i + 0.2 for i in x], model_names)
plt.ylim(0, 1)
plt.ylabel("Score")
plt.title("Comparison of Accuracy and AUC for Models")
plt.legend()
plt.show()

```
